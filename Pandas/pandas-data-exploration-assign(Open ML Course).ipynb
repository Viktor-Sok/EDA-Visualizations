**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year?**

```python
data.salary.unique()
```

```python
data["salary"] = data["salary"].map({"<=50K" : False, ">50K" : True}) # convert column to bool
```

```python
a = data["salary"] # this is not a copy but a reference to a series which consist the column in original table
a1 = data["salary"].copy() # this is a copy
type(a1)
```

```python
data[["salary", "age"]].groupby("salary").agg([np.mean, np.std]) 
```

```python
ages1 = data.loc[data["salary"] == True, "age"]
ages2 = data.loc[data['salary'] == False, 'age']
print("The average age of the rich: {0} +- {1} years, poor - {2} +- {3} years.".format(
    round(ages1.mean()), round(ages1.std(), 1),
    round(ages2.mean()), round(ages2.std(), 1)))
```

**7. Display age statistics for each race (*race* feature) and each gender (*sex* feature). Use *groupby()* and *describe()*. Find the maximum age of men of *Amer-Indian-Eskimo* race.**

**8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
data["marital-status"].unique()
```

```python
married = data[(data['marital-status'].str.contains("Married*",regex=True)) & (data['sex'] == "Male")] # choose rows by matching values in the column
married[married["salary"] == True].shape[0]
```

```python
single = data[(~data["marital-status"].isin(married["marital-status"])) & (data['sex'] == "Male")] # dataframe difference
single[single["salary"] == True].shape[0]
```

```python
data.loc[(data['sex'] == 'Male') &
     (data['marital-status'].isin(['Never-married', 
                                   'Separated', 
                                   'Divorced',
                                   'Widowed'])), 'salary'].value_counts()
```

```python
data.loc[(data['sex'] == 'Male') &
     (data['marital-status'].str.startswith('Married')), 'salary'].value_counts()
```
