## Joining/concatenating tables

### Concatenation

```python
raw_data = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}
df_a = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_a.index = [0,1,2,3,4]
df_a["subject_id"] = df_a["subject_id"].astype("int32")

raw_data = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
df_b = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_b.index = [2,3,4,5,6]
df_b["subject_id"] = df_b["subject_id"].astype("int32")
```

```python
df_a
```

```python
df_b
```

```python
# concatenation along 0 dimension (rows)
pd.concat([df_a, df_b])
```

```python
# concatenate along 1st dimension (columns)
pd.concat([df_a, df_b],axis  = 1)
```

```python
# concatenate keeping only shared indeces (inner join)
#Concat, но только с помощью атрибута "inner"
df_new_ = pd.concat([df_a, df_b],axis = 1, join='inner')
df_new_
```

### Merging (по сути join, но по любым колонкам, но можно и по индексу)

```python
# join by column subject_id, use left column: it's possible to use inner, outer, right and cross join
```

```python
# join by left index and right column
pd.merge(df_a, df_b, left_index=True, right_on = "subject_id", how = "inner")
```
